```python
from sklearn.tree import DecisionTreeClassifier
import inspect
from typing import get_type_hints

def get_parameter_info(model_func):
    """Extract parameter information from a function/model."""
    sig = inspect.signature(model_func)
    type_hints = get_type_hints(model_func)
    
    params = []
    for param_name, param in sig.parameters.items():
        # Determine type: type hint > default type > Any
        if param_name in type_hints:
            param_type = type_hints[param_name].__name__
        elif param.default != inspect.Parameter.empty:
            param_type = type(param.default).__name__
        else:
            param_type = "Any"
        
        param_info = {
            "name": param_name,
            "type": param_type,
            "default": (
                param.default if param.default != inspect.Parameter.empty 
                else None
            ),
        }
        params.append(param_info)
    
    return params


get_parameter_info(DecisionTreeClassifier.__init__)
```
