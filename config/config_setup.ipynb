```python
from models import StaticPage, ModelPage

intro = StaticPage(parameters={"link": "/dtree/introduction.md"})

# decision_tree_building =  ModelPage(component_type="train",
#                                     model_name="decision_tree", 
#                                     parameters={"whitelist": ["max_depth"]})

decision_tree_building = ModelPage(component_type="manual",                 model_name="decision_tree")

prediction_text = StaticPage(parameters={"link": "/dtree/prediction.md"})


decision_tree_inference = ModelPage(component_type="predict",
                                    model_name="decision_tree")

end = StaticPage(parameters={"link": "dtree/end.md"})

```

```python
from models import create_page_dict

page_list = [intro, decision_tree_building, prediction_text, decision_tree_inference, end]

pages = create_page_dict(page_list)

pages
```

```python
from models import Edge, EdgeNode, AndCondition, ParameterCheck

inference_condition = AndCondition(
    conditions= [
        ParameterCheck(parameter="sepal length (cm)",
                    category="predictParams",
                    comparator="=",
                    value=7.1),
        ParameterCheck(parameter="sepal width (cm)",
                    category="predictParams",
                    comparator="=",
                    value=3.0),
        ParameterCheck(parameter="petal length (cm)",
                   category="predictParams",
                   comparator="=",
                   value=5.9),
        ParameterCheck(parameter="petal width (cm)",
                    category="predictParams",
                    comparator="=",
                    value=2.1)
    ]
)

edges = [
    Edge(start=EdgeNode(local_index=0), 
         end=EdgeNode(local_index=1)),
    Edge(start=EdgeNode(local_index=1), 
         end=EdgeNode(local_index=2)),
    Edge(start=EdgeNode(local_index=2), 
         end=EdgeNode(local_index=3)),
    Edge(start=EdgeNode(local_index=3), 
         end=EdgeNode(local_index=4),
         condition=inference_condition),
    Edge(start=EdgeNode(local_index=4),
         end=EdgeNode(local_index=0))
]
```

```python
from models import Story, StoryNode, compute_object_hash


nodes = [StoryNode(index = p) for p in pages.keys()]

story = Story(name="decision_tree_classifier", 
              description="An introduction to DecisionTreeClassifiers.",
              nodes=nodes,
              edges=edges)

story
```

```python
from models import Config


config = Config(stories={story.name: story},
                pages=pages)

with open("../frontend/src/assets/config.json", "w", encoding="utf-8") as f:
    f.write(config.model_dump_json(indent=4))
```
